# Uncertainty score - output
## How to use:
1. Choose the image you want to check from a dataset and give to the function **image_to_tensor** as input the image index and dataset name. Save the output as your sample.
2. Run the function **output_certainty** with the model name and the sample name as inputs. It will return the certainty of the prediction on a scale *to be defined*.

```python
# to suppress scientific notation (exponential)
torch.set_printoptions(sci_mode=False)
```

```python
def image_to_tensor(image_index, dataset):
    image = [dataset[image_index][0]]
    # label optional, just to keep track
    label = [dataset[image_index][1]]
    # for final use, delete the 2nd object - the label (not present for "real" test image)
    return (torch.stack(image).float(), torch.tensor(label).long())
```

```python
# initial concept:
##  "high certainty" - "medium" - "low" - "very low"
##                  0.9        0.75     0.6
# literature research to support the choice needed

def output_certainty(model, sample):
    scores = prediction_scores(model, sample)
    print
    for i in scores.numpy()[0]:
        if i > 0.9:
            return "High certainty"
    for i in scores.numpy()[0]:
        if i > 0.75:
            return "Medium certainty"
    for i in scores.numpy()[0]:
        if i > 0.6:
            return "Low certainty"
    return "Very low certainty"
```
